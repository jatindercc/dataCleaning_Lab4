## **Outlier Detection** 


An Outlier is a data-item/object that deviates significantly from the rest of the (so-called normal)objects. They can be caused by measurement or execution errors. The analysis for outlier detection is referred to as outlier mining. There are many ways to detect the outliers, and the removal process is the data frame same as removing a data item from the pandaโs data frame.

#### Z-Score:
Z- Score is also called a standard score. This value/score helps to understand that how far is the data point from the mean. And after setting up a threshold value one can utilize z score values of data points to define the outliers.
<br>
Zscore = (data_point -mean) / std. deviation



```python
# Calculate Z-scores for BMI
# Z-score indicates how many standard deviations a data point is from the mean
# (value - mean) / std_deviation

bmi_mean = diabetes_df['bmi'].mean()
bmi_std = diabetes_df['bmi'].std()
diabetes_df['bmi_zscore'] = (diabetes_df['bmi'] - bmi_mean) / bmi_std

# Set threshold and filter outliers
threshold = 3
outliers = diabetes_df[np.abs(diabetes_df['bmi_zscore']) > threshold]

# Display outliers
print("Outliers based on BMI Z-score:")
print(outliers[['bmi', 'bmi_zscore', 'target']])
```

Now to define an outlier threshold value is chosen which is generally 3.0. As 99.7% of the data points lie between +/- 3 standard deviation (using Gaussian Distribution approach).

Rows where Z value is greater than 2

```python
# Calculate Z-score for BMI
bmi_mean = diabetes_df['bmi'].mean()
bmi_std = diabetes_df['bmi'].std()
diabetes_df['bmi_zscore'] = (diabetes_df['bmi'] - bmi_mean) / bmi_std

# Filter rows where Z-score > 2
z_gt_2 = diabetes_df[diabetes_df['bmi_zscore'] > 2]

# Display results
print("Rows where BMI Z-score > 2:")
print(z_gt_2[['bmi', 'bmi_zscore', 'target']])
```

#### IQR (Inter-Quartile Range)
Inter Quartile Range approach to finding the outliers is the most commonly used and most trusted approach used in the research field. <Br>
IQR = Quartile3 - Quartile1

```python
# Calculate Q1 and Q3 for BMI
Q1 = diabetes_df['bmi'].quantile(0.25)
Q3 = diabetes_df['bmi'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
iqr_outliers = diabetes_df[(diabetes_df['bmi'] < lower_bound) | (diabetes_df['bmi'] > upper_bound)]

# Display outliers
print("Outliers based on BMI using IQR:")
print(iqr_outliers[['bmi', 'target']])
```

To define the outlier base value is defined above and below datasetโs normal range namely Upper and Lower bounds, define the upper and the lower bound (1.5*IQR value is considered) :<br>
upper = Q3 + 1.5 * IQR <br>
lower = Q1 - 1.5 * IQR <br> 

```python
Q1 = diabetes_df['bmi'].quantile(0.25)
Q3 = diabetes_df['bmi'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
```

- Removing the outliers:
For removing the outlier, one must follow the same process of removing an entry from the dataset using its exact position in the dataset because in all the above methods of detecting the outliers end result is the list of all those data items that satisfy the outlier definition according to the method used.

```python
# Calculate IQR bounds
Q1 = diabetes_df['bmi'].quantile(0.25)
Q3 = diabetes_df['bmi'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
diabetes_cleaned = diabetes_df[(diabetes_df['bmi'] >= lower_bound) & (diabetes_df['bmi'] <= upper_bound)]

print(f"Original rows: {len(diabetes_df)}")
print(f"Cleaned rows: {len(diabetes_cleaned)}")
```
